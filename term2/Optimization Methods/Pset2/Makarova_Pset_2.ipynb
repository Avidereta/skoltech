### Problem 2.

```python
from cvxpy import *
import numpy as np


np.random.seed(1)

def Road_light(m, n):
    p = Variable(m)
    A = np.random.random((n,m))
    I = np.random.random(n)

    z = Variable()
    constraints = [p >= 0, abs(A * p - I) <= z]

    objective = Minimize(z)

    prob = Problem(objective, constraints)
    prob.solve(solver = "GUROBI", Method = 1) 
    
    
    return prob.status, prob.value, p.value

m = 4
n = 20

status, value, var = Road_light(m,n)

print "status of solution:", status
print "minimum of objective function = ", value
print "optimal variables \n{0}".format(var)                   

```

### Problem 3

#### (a) Formulate a linear program problem:

* $l_1, l_2$ - assemble labour
* $t_1, t_2$ - testing
* $m_1, m_2$ - expenses on materials
* $p_1, p_2$ - profit 
* $n_1, n_2$ - number of made produces per day

$$maximize: n_1(p_1 - m_1) + n_2(p_2 - m_2),$$

$$s.t. \ \ \ \ \  n_1l_1 + n_2l_2 \leq 90 $$
$$\ \ \ \ n_1t_1 + n_2t_2 \leq 80$$
$$\ \ \ \ \ m_i \geq 0, l_i \geq 0, t_i \geq 0$$
$$n_i \in N$$

According our numbers:

$$maximize: 7.8 n_1 + 7.1 n_2,$$

$$s.t. \ \ \ \ \  \frac{1}{4}n_1+ \frac{1}{3}n_2 \leq 90 $$
$$\ \ \ \ \frac{1}{8}n_1 + \frac{1}{3} n_2 \leq 80$$
$$\ \ \ \ \ n_i \in N.$$




```python
from cvxpy import *


n1 = Variable()
n2 = Variable()

constraints = [n1/4.0 + n2/3.0 <= 90, n1/8.0 + n2/3.0 <= 80, n1 >=0, n2 >= 0]

objective = Maximize(7.8 * n1 + 7.1 * n2)

prob = Problem(objective, constraints)
prob.solve(solver = "GUROBI", Method = 1) 

print "status of solution:", prob.status
print "maximum of objective function =", prob.value
print "optimal variables (n1, n2) = ({0}, {1}):".format(n1.value, n2.value)
```

#### (b1). 50 hours of overworking:
* $T_1, T_2$ - overworking at each product type 

$$maximize: 7.8 n_1 + 7.1 n_2 - 7(T_1 + T_2),$$

$$s.t. \ \ \ \ \  \frac{1}{4}n_1+ \frac{1}{3}n_2 \leq 90 + T_1$$
$$\ \ \ \ \frac{1}{8}n_1 + \frac{1}{3} n_2 \leq 80 + T_2$$ 

$$T_1 + T_2 \leq 50$$
$$\ \ \ \ \ n_i \in N$$

So, we have a linear programming problem. The main problem is that n should \in $N$ (linear integer programming problem) and we have NP- hard problem. The solution can be find ignoring the integrality constraints and obtain the so-called linear programming relaxation of the original problem. Because the linear programming problem
has fewer constraints, and therefore more options, the optimal cost will be
less than or equal to the optimal cost of the original problem. If the optimal
solution to the linear programming relaxation happens to be integer, then
it is also an optimal solution to the original problem. If it is not integer, we
can round each Xj upwards, thus obtaining a feasible, but not necessarily
optimal, solution to the original problem. 


```python
from cvxpy import *


n1 = Variable()
n2 = Variable()
T1 = Variable()
T2 = Variable()

constraints = [n1/4.0 + n2/3.0 <= 90 + T1, n1/8.0 + n2/3.0 <= 80 + T2, \
               T1 + T2 <= 50, T1 >= 0, T2 >= 0, n1 >=0, n2 >=0]

objective = Maximize(7.8 * n1 + 7.1 * n2 - 7.0 * (T1 + T2))

prob = Problem(objective, constraints)
prob.solve(solver = "GUROBI", Method = 1) 

print "status of solution:", prob.status
print "minimum of objective function =", prob.value
print "optimal variables (n1, n2) = ({0}, {1}):".format(n1.value, n2.value)
```


#### (b2). Addition discount:

On the one hand this problem can not be considered as a simple linear programming problem. On the other hand, having the initial profit

$$F = 7.8 n_1 + 7.1 n_2$$

we are aimed at finding 

$$max \{F, F + 0.1(1.2n_1 + 0.9n_2)\}$$

where for the second we have an addition constraint $F + 0.1(1.2n_1 + 0.9n_2) \geq 300.$ 
So, we can solve two linear programming problems (same as ones above) and choose one with the best result for objective function. 

```python
from cvxpy import *


n1 = Variable()
n2 = Variable()


constraints = [n1/4.0 + n2/3.0 <= 90, n1/8.0 + n2/3.0 <= 80, \
               7.8 * n1 + 7.1 * n2 + 0.1 * (1.2 * n1 + 0.9 * n2) >= 300 , n1 >=0, n2 >=0]

objective = Maximize(7.8 * n1 + 7.1 * n2 + 0.1 * (1.2 * n1 + 0.9 * n2))

prob = Problem(objective, constraints)
prob.solve(solver = "GUROBI", Method = 1) 

print "status of solution:", prob.status
print "maximum of objective function =", prob.value
print "optimal variables (n1, n2) = ({0}, {1}):".format(n1.value, n2.value)
```

The result for objective function is greater with the discount system, than with the initial system, although optimal variables are the same: 2851.2 against 2808.0.
