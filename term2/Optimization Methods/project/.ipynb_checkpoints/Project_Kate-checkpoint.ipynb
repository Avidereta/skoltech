```python
np.random.seed(1)
def generate_adj_matrix(people_number, fill_factor = 0.3):
    fill_coeff = int(people_number * fill_factor)
    adj_matrix = sp.sparse.rand(people_number,
                                people_number,
                                density = fill_factor,
                                format = 'csr',
                                random_state = 42)

    adj_matrix = adj_matrix.todense()
    adj_matrix[np.arange(people_number), np.arange(people_number)] = 0
    adj_matrix = adj_matrix - adj_matrix.T
    adj_matrix[adj_matrix < 0] = 0
    
    for i in range(adj_matrix.shape[0]):
        for j in range(adj_matrix.shape[1]):
            adj_matrix[i, j] = np.int(100 * adj_matrix[i, j])

    return csr_matrix(adj_matrix)

people_number = 5
adj_matrix = generate_adj_matrix(people_number).todense()
print adj_matrix
```

Matrix $x$ represents whether $i^{th}$ person gives money to $j^{th}$ person. Matrix $c$ represents how much money is transfered during the transaction between $i$ and $j$. Then the task is formulated in the following way:

$$
\min_{x, c} \sum_{i,j=1}^n c_{ij}x_{ij} \\
\text{s.t.:} \quad c \ge 0 \\
\sum_{i=1}^n c_{ij} = \sum_{i=1}^n u_{ij} - \sum_{j=1}^n u_{ij} \\
\sum_{j=1}^n c_{ij} = \sum_{j=1}^n u_{ij} - \sum_{i=1}^n u_{ij} \\
c_{ij} \le x_{ij}c_{ij} \quad \forall i,j = \overline{1,n}
$$

```python
import cvxpy as cvx
import gurobi

u = adj_matrix.copy()

x = cvx.Bool(people_number, people_number)
c = cvx.Variable(people_number, people_number)
        
constraints = [cvx.sum_entries(c, axis=1) == cvx.pos(np.sum(u, axis=1) - np.sum(u, axis=0).T),
               cvx.sum_entries(c, axis=0) == cvx.pos(np.sum(u, axis=0) - np.sum(u, axis=1).T),
               c <= 1000 * x,
               c >= 0]

obj = cvx.Minimize(cvx.sum_entries(x))

prob = cvx.Problem(obj, constraints)
prob.solve(solver='GUROBI')
print 'optimal value:', prob.value
print 'optimal x: \n', x.value
print 'optimal ff: \n', ff.value
```
